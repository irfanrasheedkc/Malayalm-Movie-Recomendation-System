```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('/content/movie_details.csv')
```

```python
# Function to clean and split names
def clean_and_split(names):
    cleaned_names = names.replace('[', '').replace(']', '').replace(',', ' ').replace("'", "")
    return cleaned_names

# Apply the function to the 'cast' column
movies_data['cast'] = movies_data['cast'].apply(clean_and_split)
```

```python
movies_data.head()
```

```python
import nltk

nltk.download('averaged_perceptron_tagger')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

# # Example sentence
# sentence = "Natural language processing is a field of artificial intelligence that focuses on the interaction between computers and humans."
def get_nouns(sentence):
  if 'Add Plot' in sentence:
    return 'null'
  # Tokenize the sentence into words
  words = word_tokenize(sentence)

  # Filter out stopwords
  stop_words = set(stopwords.words("english"))
  filtered_words = [word for word in words if word.lower() not in stop_words]

  # Tag words with part-of-speech
  pos_tags = nltk.pos_tag(filtered_words)

  # Extract nouns (subjects)
  subjects = [word for word, pos in pos_tags if pos.startswith('NN')]

  print(subjects)
  return ' '.join(subjects)

# Apply the function to the 'cast' column
movies_data['description'] = movies_data['description'].apply(get_nouns)
```

```python
movies_data.head(20)
```

```python
# selecting the relevant features for recommendation

selected_features = ['genre','description','cast','director']
print(selected_features)
```

```python
for feature in selected_features:
    movies_data[feature] = movies_data[feature].apply(lambda x: '' if x is None or (isinstance(x, str) and x.lower() == 'null') else x)
movies_data.fillna('', inplace=True)
```

```python
movies_data.head()
```

```python
# Replace 'Add Plot' with an empty string in the 'description' column
movies_data['description'] = movies_data['description'].replace('Add Plot', '')

# Now, if 'Add Plot' existed in the 'description' column, it has been replaced with an empty string.
```

```python
# combining all the 5 selected features
combined_features = movies_data['genre']+' '+movies_data['cast']
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
# Load the similarity matrix from the file
similarity= np.load('similarity.npy')
print('Similarity matrix loaded successfully.')

# Loading the list of movie titles from the file
list_of_all_titles = np.load('list_of_all_titles.npy')
print('List of movie titles loaded successfully.')

def get_movie_details():
  # getting the movie name from the user
  movie_name = input(' Enter your favourite movie name : ')

  # creating a list with all the movie names given in the dataset
  list_of_all_titles = movies_data['title'].tolist()
  print(list_of_all_titles)

  # finding the close match for the movie name given by the user
  find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
  print(find_close_match)

  close_match = find_close_match[0]
  print(close_match)

  index_of_the_movie = movies_data.loc[movies_data['title'] == close_match].index[0]
  print(index_of_the_movie)

  similarity_score = list(enumerate(similarity[index_of_the_movie]))
  print(similarity_score)

  # sorting the movies based on their similarity score

  sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
  print(sorted_similar_movies)

  # print the name of similar movies based on the index

  print('Movies suggested for you : \n')

  i = 1

  for movie in sorted_similar_movies:
      index = int(movie[0])
      if i <= 5:  # Print top 5 similar movies
          title_from_index = movies_data.iloc[index]['title']
          similarity_score = movie[1]
          print(i, '.', title_from_index, ' (Similarity Score:', similarity_score, ')')
          i += 1

get_movie_details()
```
